# Convert Jupyter Notebook to HTML

After you have completed the lab exercises, as a part of your submission, you need to first convert the jupyter notebook (.ipynb) file to HTML file. The following instructions will guide you for this conversion.

For this lab, we are trying to convert the file "Lab_1_Intro_to_Jupyter_Table.ipynb" to HTML. This file is in your Google Drive. Since this current notebook (Convert_Jupyter_Notebook_To_HTML.ipynb) is a completely new notebook, we need to first mount the Google Drive here so that the drive is accessible from this notebook as well. Run the following command and complete the authorization process as you did in the other notebook.

```
from google.colab import drive
drive.mount('/content/drive')
```

After you successfully mount the drive, you are in the /content directory. You need to find out where the file, that you want to convert, is. This "content" directory has a sub-directory called "drive". Inside this sub-directory "drive", there is another sub-directory called "My Drive/ which is your actual Google Drive. In this case, the file (Lab1YourLasNemeYourFirstName.ipynb) which we want to convert into HTML is inside the "Colab Notebooks" directory in "My Drive" directory.

"ls" is a unix command to list the files and sub-directories inside a directory. For instance, if you run "ls /content/drive", it will list all the files and directories inside the drive directory. Generally, when running command-line commands from Jupyter Notebook, we add an "!" at the beginning to inform the interpreter that this is a system command.

```
!ls
```

```
!ls /content/drive
```

# Because there is a space between "My" and "Drive", we need to add a back slash in Jupyter Notebook between the two words.

```
!ls /content/drive/My\ Drive
```

## You can see the files under the "Colab Notebooks" directory, which should include your Lab1 Jupyter Notebook.

```
!ls /content/drive/My\ Drive/Colab\ Notebooks/
```

So, the file structure is as below:

- / 
  - content
    - drive
      - My Drive
        - Colab Notebooks
          - Lab1YourLastNameYourFistName.ipynb

Now, that we know where the file we want to convert is, we'll use the "jupyter nbconvert" command and the absolute file path to convert it into html. You need to replace "Lab1YenJohn.ipynb" in the following command with the name of your Lab1 file using your name: 

```
!jupyter nbconvert --to html /content/drive/My\ Drive/Colab\ Notebooks/Lab11WeiHaichen.ipynb
```

So, a new html file with the same name ("Lab_1_Intro_to_Jupyter_Table") but different extension (.html) will be created on the same directory. We can check this using the "ls" command.

```
!ls /content/drive/My\ Drive/Colab\ Notebooks/
```

You can now go to your google drive, find your Lab1 .html file under Colab Notebooks directory. Download this html file and submit it through Canvas.
